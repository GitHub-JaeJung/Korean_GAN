```python
import numpy as np

import pandas as pd

import os #파일이 몇 개 있는지(파일명) 등을 os 기능을 사용해요

import cv2 as cv #open cv 이미지 파일 읽기

from sklearn import utils # sklearn의 shuffle과 같은 기능 사용

from tqdm.notebook import tqdm
```

```python
train_dir = './train/img/'
```

```python
def labeling(img):

    class_name = img.split('.')[0]

    if class_name == 'da':
        return 0

    elif class_name == 'ram':
        return 1
    
    elif class_name == 'jwi':
        return 2
    elif class_name == 'heon':
        return 3
    elif class_name == 'ches':
        return 4
    elif class_name == 'ba':
        return 5
    elif class_name == 'kwi':
        return 6
    elif class_name == 'e_':
        return 7
    elif class_name == 'ta':
        return 8
    elif class_name == 'go':
        return 9
    elif class_name == 'fa':
        return 10    
```

```python
x_data = list()
t_data = list()
```

```python
# print(os.listdir(train_dir))

for img in tqdm(os.listdir(train_dir), total=len(os.listdir(train_dir)), position=0, leave=True):
    label_data = os.listdir(train_dir) # 현재 파일의 label을 알아내요
    img_path = os.path.join(train_dir,img) # 파일 경로와 우리의 이미지 data를 이용해 경로 지정
    img_data = cv.resize(cv.imread(img_path, cv.IMREAD_GRAYSCALE), (32,32))
    x_data.append(img_data.ravel(order = 'F'))
    t_data.append(label_data)
```

```python
x_df = pd.DataFrame(x_data)
# t_df = pd.DataFrame({ 'label' : t_data }) # dictionary의 key 값이 column 명 # label

    
display(x_df)
```

```python
df = pd.merge(t_df, x_df, left_index=True, right_index=True) # index 이용 merge

display(df)
```

```python
shuffled_df = utils.shuffle(df) # sklearn의 shuffle 기능
```

```python
shuffled_df.to_csv('./train/train.csv', index=False)
```
